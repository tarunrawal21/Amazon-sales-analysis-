```python
AS['Region'] = AS['Region'].astype('category')
AS['Country'] = AS['Country'].astype('category')
AS['Item Type'] = AS['Item Type'].astype('category')
AS['Item Type'] = AS['Item Type'].astype('category')
AS['Sales Channel'] = AS['Sales Channel'].astype('category')
AS['Order Priority'] = AS['Order Priority'].astype('category')
AS['Order Date'] = pd.to_datetime(AS['Order Date'])
AS['Ship Date'] = pd.to_datetime(AS['Ship Date'])
```

# **Statistical analysis**

```python
corr = AS[['Total Revenue', 'Units Sold']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Total Revenue and Units sold are very correlated to each othre

# **Data visualization**

```python
avg_revenue_by_region = AS.groupby('Region')['Total Revenue'].mean()
print(avg_revenue_by_region)
```

```python
avg_revenue_by_region.plot(kind='bar', color = 'orange', edgecolor= 'black')
plt.title('Average Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Average Revenue')
plt.show()
```

```python
avg_revenue_by_country = AS.groupby('Country')['Total Revenue'].mean()
print(avg_revenue_by_country)
```

```python
avg_revenue_by_country.plot(kind='bar', color = 'red', edgecolor= 'black', figsize=(15, 6))
plt.title('Average Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Average Revenue')
plt.show()
```

```python
avg_units_price_by_region = AS.groupby('Region')['Unit Price'].mean()
print(avg_units_price_by_region)
```

```python
avg_units_price_by_region.plot(kind='bar', color = 'green', edgecolor= 'black')
plt.title('Average Unit Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Unit Price')
plt.show()
```

```python
avg_total_profit_by_region = AS.groupby('Region')['Total Profit'].mean()
print(avg_total_profit_by_region)
```

```python
region = avg_total_profit_by_region.nlargest()
region.plot(kind='line', color = 'blue', marker = '*', markerfacecolor= 'black')
plt.title('Average Total Profit by Region')
plt.figsize=(4, 4)
plt.xticks(rotation=45)
plt.xlabel('Region')
plt.ylabel('Average Total Profit')
plt.show()
```
